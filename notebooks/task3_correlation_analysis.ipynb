```python
import pandas as pd
import matplotlib.pyplot as plt

# Load GOOG stock data
goog = pd.read_csv("data/raw/GOOG.csv")

# Prepare stock data
goog["Date"] = pd.to_datetime(goog["Date"])
goog["Daily_Return"] = goog["Close"].pct_change()
goog["date_only"] = goog["Date"].dt.date

# Create a small sample of daily news counts
daily_news = pd.DataFrame({
    "date_only": pd.to_datetime([
        "2011-04-28",
        "2011-04-29",
        "2011-05-02",
        "2011-05-03",
        "2011-05-05",
        "2011-05-06",
        "2011-05-09",
        "2011-05-11"
    ]).date,
    "news_count": [3, 2, 9, 3, 3, 3, 3, 3]
})

# Merge stock returns with news counts
merged = pd.merge(
    goog[["date_only", "Daily_Return"]],
    daily_news,
    on="date_only",
    how="inner"
)

# Remove missing values
merged = merged.dropna()

# Calculate correlation
correlation = merged["news_count"].corr(merged["Daily_Return"])
print("Correlation between news count and daily returns:", correlation)

# Display merged data
print("\nMerged Data:")
print(merged)

# Create scatter plot
plt.figure(figsize=(8, 6))
plt.scatter(merged["news_count"], merged["Daily_Return"], alpha=0.7)
plt.title("News Volume vs Daily Returns (GOOG)")
plt.xlabel("Number of News Articles")
plt.ylabel("Daily Return")
plt.grid(True)
plt.show()
```
